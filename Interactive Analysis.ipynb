```python
from geopy import distance

college_geo = []
chickfila_geo = []
with open('colleges_locations.txt') as colleges, open('chickfila_locations.txt', encoding='utf-8') as chickfilas:
    college_geo = [c.split(':') for c in colleges.read().splitlines()]
    chickfila_geo = [c.split(':') for c in chickfilas.read().splitlines()]

def dist_between(geo1, geo2):
    return distance.distance(geo1, geo2).miles

def dist_lis(geo1, lis):
    return [[dist_between(geo1, eval(loc)), loc, name] for name, loc in lis]

def find_college(name):
    for college_name, loc in college_geo:
        if name.strip().lower() in college_name.strip().lower():
            return [college_name, loc]
    raise FileNotFoundError
```

```python
# Find the college by name
col = find_college('Massachusetts Institute of Technology')
col
```

```python
# Find distance to all chickfilas and sort
col_loc = dist_lis(eval(col[1]), chickfila_geo)
col_loc.sort()
col_loc
```
